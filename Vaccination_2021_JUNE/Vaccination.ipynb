## Vaccine doses per municipality (includes state, vaccine company, and dose 1/2 count)

```python
BRAZIL_POPULATION_23JUN = 214054369
doses_aplicadas = pd.read_csv('doses_por_municipio.csv')
print('Total doses administered: %d' % sum(doses_aplicadas['Doses Aplicadas']))
print('Proportion of popluation with at least 1 dose: %f' % (sum(doses_aplicadas['Dose 1']) / BRAZIL_POPULATION_23JUN))
print('\nTable overview:')
doses_aplicadas.head()
```
